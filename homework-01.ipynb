```python
# import relative package
import urllib.request
import csv
import pandas as pd
```

```python
# download data & to pandas
url = 'https://data.nhi.gov.tw/resource/mask/maskdata.csv'
urllib.request.urlretrieve(url, './data/maskdata.csv')
with open('./data/maskdata.csv') as csvfile:
    csvreader = csv.reader(csvfile)
    header = next(csvreader)
    rows = list(csv.reader(csvfile))
df_mask = pd.DataFrame(data=rows, columns=header)
```

```python
df_mask['countyname'] = df_mask['醫事機構地址'].apply(lambda x: x[:3])
replace_words = {
    "臺中縣":"臺中市"
}
df_mask.countyname = df_mask.countyname.replace(regex = replace_words)
df_mask['成人口罩剩餘數'] = df_mask['成人口罩剩餘數'].astype('int')
df_mask['兒童口罩剩餘數'] = df_mask['兒童口罩剩餘數'].astype('int')
df_mask['mask_cnt'] = df_mask['成人口罩剩餘數'] + df_mask['兒童口罩剩餘數']
```

```python
df_mask.groupby(['countyname'])['mask_cnt'].sum().reset_index()
```
